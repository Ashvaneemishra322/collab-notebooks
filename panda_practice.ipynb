```
data = [
    ['Aman' , 'Ravi' , 'Sita' , 'karan' , 'marry'],
    [45, 67, 82, 30, 55]
]
sam = pd.DataFrame(data).T
sam.columns = ['Name' , 'Marks']
```

```
sam
```

```
# Create a new column Grade

# Rules:

# ≥ 80 → A

# ≥ 60 → B

# ≥ 40 → C

# < 40 → Fail

###### create a function #######

def grade(mark):

  if (mark >= 80):
    return 'A'
  elif (mark >= 60):
    return 'B'
  elif (mark >= 40):
    return 'C'
  else:
    return 'Fail'

```

```
sam['Grade'] = sam["Marks"].apply(grade)
```

```
sam
```
