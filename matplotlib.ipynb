```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

** Matpltlib basics **

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

subplots in matplotlib


```python
plt.subplot(2,2,1)
plt.plot(x)
plt.subplot(2,2,2)
plt.plot(y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x**2 , y**2)
```

Object oriented way

```python
fig = plt.figure()
axes1  = fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2  = fig.add_axes([0.3,0.4,0.3,0.3])
axes2.plot(x,y)
```

Types of plot

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

Saving plot

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basic_plot.png")
```
